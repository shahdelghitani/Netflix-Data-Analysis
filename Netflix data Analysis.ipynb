### Tv Shows vs year

```python
movies = data[data['type'] == 'Movie']
tv_shows = data[data['type'] == 'TV Show']

movies_count = movies.groupby('release_year').size().reset_index(name='count')
tv_shows_count = tv_shows.groupby('release_year').size().reset_index(name='count')

plt.figure(figsize=(10, 6))
plt.plot(movies_count['release_year'], movies_count['count'], marker='o', linestyle='-', color='b', label='Movies')
plt.title('Number of Movies Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()

plt.figure(figsize=(10, 6))
plt.plot(tv_shows_count['release_year'], tv_shows_count['count'], marker='o', linestyle='-', color='g', label='TV Shows')
plt.title('Number of TV Shows Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.tight_layout()


plt.show()
```
